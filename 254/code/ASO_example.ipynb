```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn a

# Define the parameters of the first normal distribution
mu1 = 0
sigma1 = np.sqrt(5)

# Define the parameters of the second normal distribution
mu2 = -2
sigma2 = np.sqrt(0.5)

# Create an array of x values to evaluate the CDFs
x = np.linspace(-5, 5, 1000)

# Evaluate the CDFs of the two normal distributions
cdf1 = norm.cdf(x, mu1, sigma1)
cdf2 = norm.cdf(x, mu2, sigma2)

# Plot the CDFs
plt.plot(x, cdf1, label='N(0, 5)')
plt.plot(x, cdf2, label='N(-2, 0.5)')

intersection = cdf2 - cdf1
plt.fill_between(x, intersection, alpha=0.5, where=(intersection > -5))


plt.legend()
plt.xlabel('x')
plt.ylabel('CDF')
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set the style of the plot
sns.set_style("whitegrid")

# Define the parameters of the first normal distribution
mu1 = 0
sigma1 = np.sqrt(5)

# Define the parameters of the second normal distribution
mu2 = -2
sigma2 = np.sqrt(0.5)

# Create an array of x values to evaluate the CDFs
x = np.linspace(-5, 5, 1000)

# Evaluate the CDFs of the two normal distributions
cdf1 = norm.cdf(x, mu1, sigma1)
cdf2 = norm.cdf(x, mu2, sigma2)


# Create the plot
fig, ax = plt.subplots()

# Shade the region of intersection
ax.fill_between(x, cdf2, cdf1, alpha=0.5, where=(intersection < 0), interpolate=True)

# Plot the CDFs
sns.lineplot(x=x, y=cdf1, label='N(0, 5)', ax=ax)
sns.lineplot(x=x, y=cdf2, label='N(-2, 0.5)', ax=ax)

# Set the axis labels and title
ax.set_xlabel('x')
ax.set_ylabel('CDF')
ax.set_title('Intersection of two normal distributions')

# Show the plot
plt.show()
```
